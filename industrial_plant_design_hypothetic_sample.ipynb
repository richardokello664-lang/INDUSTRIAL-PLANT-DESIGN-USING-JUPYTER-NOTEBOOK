## 3. Basic process and plant calculations

```python

# Simplified design assumptions
production_rate_tph = 12.0
operating_hours_per_day = 20
operating_days_per_year = 330
plant_efficiency = 0.92
product_density_t_m3 = 0.85
electricity_cost_ksh_per_kwh = 30

daily_production_t = production_rate_tph * operating_hours_per_day * plant_efficiency
annual_production_t = daily_production_t * operating_days_per_year
daily_product_volume_m3 = daily_production_t / product_density_t_m3

total_connected_power_kw = equipment["power_kw"].sum()
daily_energy_kwh = total_connected_power_kw * operating_hours_per_day
annual_energy_kwh = daily_energy_kwh * operating_days_per_year
annual_electricity_cost_ksh = annual_energy_kwh * electricity_cost_ksh_per_kwh

summary = pd.DataFrame({
    "Metric": [
        "Effective daily production",
        "Annual production",
        "Daily product volume",
        "Total connected power",
        "Daily electrical energy",
        "Annual electrical energy",
        "Estimated annual electricity cost"
    ],
    "Value": [
        daily_production_t,
        annual_production_t,
        daily_product_volume_m3,
        total_connected_power_kw,
        daily_energy_kwh,
        annual_energy_kwh,
        annual_electricity_cost_ksh
    ],
    "Unit": [
        "tonnes/day",
        "tonnes/year",
        "m³/day",
        "kW",
        "kWh/day",
        "kWh/year",
        "KSh/year"
    ]
})

summary
```

## 4. Storage adequacy check

```python

product_tank_capacity_m3 = equipment.loc[
    equipment["tag"] == "TK-102", "capacity_m3"
].iloc[0]

storage_days = product_tank_capacity_m3 / daily_product_volume_m3

print(f"Product tank capacity: {product_tank_capacity_m3:.1f} m³")
print(f"Estimated daily product volume: {daily_product_volume_m3:.1f} m³/day")
print(f"Available product storage: {storage_days:.2f} days")

if storage_days >= 1:
    print("Result: Storage is adequate for at least one day of production.")
else:
    print("Result: Consider increasing product storage capacity.")
```

## 6. Approximate pipe lengths and pressure-drop indicator

```python

equipment_lookup = equipment.set_index("tag")

pipe_results = []

for _, pipe in pipes.iterrows():
    start = equipment_lookup.loc[pipe["from_tag"]]
    end = equipment_lookup.loc[pipe["to_tag"]]

    dx = end["x"] - start["x"]
    dy = end["y"] - start["y"]
    dz = (end["z"] + end["height"]/2) - (start["z"] + start["height"]/2)

    straight_length = math.sqrt(dx**2 + dy**2 + dz**2)
    design_length = straight_length * 1.20  # 20% allowance for routing and fittings

    # Educational pressure-drop indicator, not a full hydraulic calculation
    pressure_drop_index = design_length / (pipe["diameter_mm"] / 1000) ** 5

    pipe_results.append({
        "line": pipe["line"],
        "service": pipe["service"],
        "diameter_mm": pipe["diameter_mm"],
        "straight_length_m": round(straight_length, 2),
        "estimated_design_length_m": round(design_length, 2),
        "relative_pressure_drop_index": round(pressure_drop_index, 0)
    })

pipe_analysis = pd.DataFrame(pipe_results)
pipe_analysis
```


## 7. Interactive 3D industrial plant model

Use your mouse to:

- Rotate the plant
- Zoom in and out
- Pan across the layout
- Hover over equipment to view engineering details
- Toggle equipment, pipes and safety zones from the legend


```python

def add_box(fig, x, y, z, length, width, height, name, hover_text, opacity=0.85):
    x0, x1 = x - length/2, x + length/2
    y0, y1 = y - width/2, y + width/2
    z0, z1 = z, z + height

    vertices_x = [x0, x1, x1, x0, x0, x1, x1, x0]
    vertices_y = [y0, y0, y1, y1, y0, y0, y1, y1]
    vertices_z = [z0, z0, z0, z0, z1, z1, z1, z1]

    i = [0, 0, 0, 1, 1, 2, 4, 4, 5, 3, 2, 6]
    j = [1, 2, 4, 2, 5, 3, 5, 6, 6, 7, 3, 7]
    k = [2, 4, 5, 5, 6, 7, 6, 7, 1, 4, 7, 3]

    fig.add_trace(go.Mesh3d(
        x=vertices_x,
        y=vertices_y,
        z=vertices_z,
        i=i, j=j, k=k,
        name=name,
        hovertext=hover_text,
        hoverinfo="text",
        opacity=opacity,
        flatshading=True,
        showlegend=True
    ))


def add_vertical_vessel(fig, x, y, z, diameter, height, name, hover_text):
    theta = np.linspace(0, 2*np.pi, 50)
    z_values = np.linspace(z, z + height, 30)
    theta_grid, z_grid = np.meshgrid(theta, z_values)

    radius = diameter / 2
    x_grid = x + radius * np.cos(theta_grid)
    y_grid = y + radius * np.sin(theta_grid)

    fig.add_trace(go.Surface(
        x=x_grid,
        y=y_grid,
        z=z_grid,
        name=name,
        hovertext=hover_text,
        hoverinfo="text",
        showscale=False,
        opacity=0.90,
        showlegend=True
    ))

    # Top head
    fig.add_trace(go.Mesh3d(
        x=x + radius * np.cos(theta),
        y=y + radius * np.sin(theta),
        z=np.full_like(theta, z + height),
        name=f"{name} top",
        hoverinfo="skip",
        opacity=0.90,
        showlegend=False
    ))


fig = go.Figure()

# Plant floor
floor_x = [0, 65, 65, 0]
floor_y = [0, 0, 35, 35]
floor_z = [0, 0, 0, 0]

fig.add_trace(go.Mesh3d(
    x=floor_x,
    y=floor_y,
    z=floor_z,
    i=[0, 0],
    j=[1, 2],
    k=[2, 3],
    name="Plant floor",
    opacity=0.25,
    hoverinfo="skip",
    showlegend=True
))

# Equipment
for _, row in equipment.iterrows():
    hover = (
        f"<b>{row['tag']} — {row['name']}</b><br>"
        f"Type: {row['type']}<br>"
        f"Position: ({row['x']}, {row['y']}, {row['z']}) m<br>"
        f"Dimensions: {row['length']} × {row['width']} × {row['height']} m<br>"
        f"Power: {row['power_kw']} kW<br>"
        f"Capacity: {row['capacity_m3']} m³"
    )

    if row["type"] in ["Vertical Tank", "Vertical Reactor", "Vertical Vessel"]:
        add_vertical_vessel(
            fig,
            row["x"], row["y"], row["z"],
            row["length"], row["height"],
            row["tag"],
            hover
        )
    else:
        add_box(
            fig,
            row["x"], row["y"], row["z"],
            row["length"], row["width"], row["height"],
            row["tag"],
            hover
        )

# Pipe routes
for _, pipe in pipes.iterrows():
    start = equipment_lookup.loc[pipe["from_tag"]]
    end = equipment_lookup.loc[pipe["to_tag"]]

    start_z = start["z"] + start["height"] * 0.55
    end_z = end["z"] + end["height"] * 0.55
    elevated_z = max(start_z, end_z, 3.5)

    route_x = [start["x"], start["x"], end["x"], end["x"]]
    route_y = [start["y"], start["y"], end["y"], end["y"]]
    route_z = [start_z, elevated_z, elevated_z, end_z]

    fig.add_trace(go.Scatter3d(
        x=route_x,
        y=route_y,
        z=route_z,
        mode="lines+markers",
        line={"width": max(3, pipe["diameter_mm"] / 18)},
        marker={"size": 3},
        name=pipe["line"],
        text=(
            f"{pipe['line']}<br>"
            f"{pipe['service']}<br>"
            f"Diameter: {pipe['diameter_mm']} mm"
        ),
        hoverinfo="text"
    ))

# Safety-clearance circles
theta = np.linspace(0, 2*np.pi, 100)

for _, row in equipment.iterrows():
    if row["safety_radius_m"] > 0:
        fig.add_trace(go.Scatter3d(
            x=row["x"] + row["safety_radius_m"] * np.cos(theta),
            y=row["y"] + row["safety_radius_m"] * np.sin(theta),
            z=np.full_like(theta, 0.08),
            mode="lines",
            line={"width": 3, "dash": "dash"},
            name=f"{row['tag']} safety zone",
            hovertext=f"{row['tag']} safety radius: {row['safety_radius_m']} m",
            hoverinfo="text",
            visible="legendonly"
        ))

# Equipment labels
fig.add_trace(go.Scatter3d(
    x=equipment["x"],
    y=equipment["y"],
    z=equipment["z"] + equipment["height"] + 1,
    mode="text",
    text=equipment["tag"],
    textposition="top center",
    name="Equipment labels",
    hoverinfo="skip"
))

fig.update_layout(
    title={
        "text": "Interactive 3D Industrial Process Plant",
        "x": 0.5
    },
    scene={
        "xaxis_title": "Plant length (m)",
        "yaxis_title": "Plant width (m)",
        "zaxis_title": "Height (m)",
        "aspectmode": "data",
        "camera": {
            "eye": {"x": 1.7, "y": 1.7, "z": 1.25}
        }
    },
    height=750,
    margin={"l": 0, "r": 0, "b": 0, "t": 60},
    legend={
        "itemsizing": "constant"
    }
)

fig.show()
```

## 9. Simple plant-layout safety check

```python

safety_checks = []

for i, item_a in equipment.iterrows():
    for j, item_b in equipment.iterrows():
        if j <= i:
            continue

        centre_distance = math.sqrt(
            (item_b["x"] - item_a["x"])**2 +
            (item_b["y"] - item_a["y"])**2
        )

        required_distance = item_a["safety_radius_m"] + item_b["safety_radius_m"]
        clearance = centre_distance - required_distance

        safety_checks.append({
            "equipment_a": item_a["tag"],
            "equipment_b": item_b["tag"],
            "centre_distance_m": round(centre_distance, 2),
            "combined_safety_radius_m": round(required_distance, 2),
            "clearance_margin_m": round(clearance, 2),
            "status": "PASS" if clearance >= 0 else "REVIEW"
        })

safety_report = pd.DataFrame(safety_checks)
safety_report.sort_values("clearance_margin_m").reset_index(drop=True)
```
